```python
NewYork_df_new = NewYork_df.copy()

NewYork_df_new['DateTime'] = NewYork_df[['Year', 'Month', 'Day', 'Hour']].apply(lambda s : datetime.datetime(*s),axis = 1)
NewYork_df_new = NewYork_df_new.drop(columns = ['Year','Minute'])
NewYork_df_new['DateTime'] = pd.to_datetime(NewYork_df_new['DateTime']) 
NewYork_df_new.set_index(['DateTime'], drop=True, inplace=True)
NewYork_df_new.head(15)
```

## LTSM 24 hour ahead

```python
def df_to_X_y_24(df, window_size):
    df_as_np = df.to_numpy()
    X = []
    y = []
    for i in range(0, df.shape[0]-window_size-24):
        row = [[a] for a in df_as_np[i:i+window_size]]
        X.append(row)
        label = df_as_np[i+window_size+24,3]
        y.append(label)
    return np.array(X), np.array(y)
```

```python
window_size = 24
X,y = df_to_X_y_24(NewYork_df_new, 24)
X.shape, y.shape
```

```python
m,n = X.shape[::3]
X = np.rollaxis(X,3,1).reshape(m,-1,n)
X.shape
```

```python
X_train, X_test, Y_train, Y_test = X[:17544],X[17544:], y[:17544], y[17544:] # Testing: from 30/12/2017 - 31/ 12/ 2018
print('Shape of X_train set:',X_train.shape) 
print('Shape of Y_train set:',Y_train.shape)
print('Shape of X_test set:',X_test.shape)
print('Shape of Y_test set:',Y_test.shape)
Y_test[:15]
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import *
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam

model2 = Sequential()
model2.add(InputLayer((24,11)))
model2.add(LSTM(50))
#model2.add(Dropout(0.5))
model2.add(Dense(50,'relu'))
#model2.add(Dropout(0.5))
model2.add(Dense(10,'relu'))
model2.add(Dense(1,'linear'))

model2.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.000001)
cp = ModelCheckpoint('model2/', save_best_only=True)
model2.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.001), metrics=RootMeanSquaredError())
history = model2.fit(X_train, Y_train, epochs=30, batch_size=128, validation_split=0.2, callbacks=[reduce_lr, cp])
```

#### Comparision between Actual and Train predictions - LSTM

```python
train_predictions = (model2.predict(X_train).flatten())
#Y_train=np.exp(Y_train)-1
MEDford_df_new = pd.DataFrame(NewYork_df_new)
train_results = pd.DataFrame(data = {'Train prediction':train_predictions,'Actual':Y_train }, index = MEDford_df_new.index[:17544])
train_results
```

```python
plt.plot(train_results['Train prediction'][300:500], label = 'Predicted')
plt.plot(train_results['Actual'][300:500], label = 'Actual')
plt.xlabel('Date')
plt.ylabel('GHI ($w/m^2$)')
plt.title('LSTM: Training vs Actuals')
_=plt.xticks(rotation = 90)
plt.legend()
```

#### Comparision between Actual and Test predictions - LSTM

```python
test_predictions = (model2.predict(X_test).flatten())
test_results = pd.DataFrame(data={'Test prediction': test_predictions, 'Actual': Y_test},index=MEDford_df_new.index[17544:17544+8712])
test_results[:15]
```

#### Testing Prediction vs Actuals - LSTM

```python
plt.plot(test_results['Test prediction'][400:600],label = 'predicted')
plt.plot(test_results['Actual'][400:600], label = 'Actual')
plt.xlabel('Date')
plt.ylabel('GHI ($w/m^2$)')
plt.title('LSTM: Testing Prediction vs Actuals')
_=plt.xticks(rotation = 90)
plt.legend()
```

#### Training and Testing scores of LSTM 

```python
from sklearn.metrics import r2_score
print('Training score:', r2_score(Y_train, train_results['Train prediction'] ))
print('Testing score:', r2_score(test_results['Actual'], test_results['Test prediction']))
test_results.corr()['Test prediction']
```
