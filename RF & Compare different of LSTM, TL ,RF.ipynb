```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

### Data Processing

```python
df = pd.read_csv("New York Solar1.csv") 
df
```

```python
df = df.dropna()
df
```

```python
NewYorkdf = pd.read_csv("New York Solar1.csv") 
NewYorkdf
```

```python
NewYorkdf = NewYorkdf.dropna()
NewYorkdf
```

### Correlatino betweeen GHI and other features

```python
NewYorkdf_new = NewYorkdf.copy()
NewYorkdf_new['DateTime'] = NewYorkdf[['Year', 'Month', 'Day', 'Hour']].apply(lambda s : datetime.datetime(*s),axis = 1)
NewYorkdf_new = NewYorkdf_new.drop(columns = ['Year','Minute'])
NewYorkdf_new['DateTime'] = pd.to_datetime(NewYorkdf_new['DateTime']) 
NewYorkdf_new.set_index(['DateTime'], drop=True, inplace=True)
NewYorkdf_new.head(15)
```

## Model training and testing 

#### Split dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV

# Features
X = NewYorkdf.copy()
X = X.drop(columns = ['GHI'])
X.index = NewYorkdf_new.index

# GHI
y = NewYorkdf_new['GHI']

X_train, X_test, Y_train, Y_test = X[:17544],X[17544:], y[:17544], y[17544:]
```

#### Grid Search to optain the best parameters

```python
param_grid = {
    'max_depth': [5, 7, 9, 11, 15, 50],
    'n_estimators': [60, 80, 100 , 120, 140, 160, 180, 200, 220] }

# Create Randfom forest model
rf = RandomForestRegressor()

# Grid search of parameters, using 3 fold cross validation, 
# search across different combinations, and use all available cores
RF_model = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
# fit model 
%time RF_model.fit(X_train,Y_train)

print(RF_model.best_params_)

# Make Prediction 
y_pred = RF_model.predict(X_test)
```

#### Comparision between Actual and Train predictions - Random Forest

```python
train_predictions = RF_model.predict(X_train).flatten()
train_results = pd.DataFrame(data = {'Train prediction':train_predictions,'Actual':Y_train }, index = NewYorkdf_new.index[:17544])
train_results
```

```python
plt.plot(train_results['Train prediction'][300:500], label = 'Predicted')
plt.plot(train_results['Actual'][300:500], label = 'Actual')
plt.xlabel('Date')
plt.ylabel('GHI ($w/m^2$)')
plt.title('Random Forest: Training vs Actuals')
_=plt.xticks(rotation = 90)
plt.legend()
```

#### Comparision between Actual and Test predictions - Random Forest

```python
Y_test.index = y[len(Y_train):].index
RF_results= pd.DataFrame(data = {'Test prediction':y_pred,'Actual':Y_test }, index =y[len(Y_train):].index)
RF_results
```

```python
plt.plot(RF_results['Test prediction'][400:600], label = 'Predicted')
plt.plot(RF_results['Actual'][400:600], label = 'Actual')
plt.xlabel('Date')
plt.ylabel('GHI ($w/m^2$)')
plt.title('Random Forest: Test vs Actuals')
_=plt.xticks(rotation = 90)
plt.legend()
```

#### Training and Testing scores of Random Forest 

```python
from sklearn.metrics import r2_score
print('Training score:', r2_score(Y_train, train_results['Train prediction'] ))
print('Testing score:', r2_score(RF_results['Actual'], RF_results['Test prediction']))
RF_results.corr()['Test prediction']
```
